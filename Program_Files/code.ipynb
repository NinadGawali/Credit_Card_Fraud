## Distributions and Feature Scaling:

Understanding the distributions of features helps us assess how skewed they are and allows for a better understanding of the dataset. In this phase, we will scale the `Time` and `Amount` columns to align them with the other numerical features (V1-V28).

#### Steps:
1. **Visualizing Distributions**: By visualizing the distributions, we can identify skewness and the nature of the features.
2. **Scaling Features**: We will use `StandardScaler` to standardize the `Time` and `Amount` columns. Standardization helps in normalizing the feature values, making them suitable for machine learning algorithms.


```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale 'Amount' and 'Time'
df['scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['scaled_time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))

# Drop the original columns
df.drop(['Time', 'Amount'], axis=1, inplace=True)

# Insert scaled columns back into the DataFrame
df.insert(0, 'scaled_amount', df.pop('scaled_amount'))
df.insert(1, 'scaled_time', df.pop('scaled_time'))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

scaled_amount_val = df['scaled_amount'].values
scaled_time_val = df['scaled_time'].values

sns.distplot(scaled_amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount post Scaling', fontsize=14)
ax[0].set_xlim([min(scaled_amount_val), max(scaled_amount_val)])

sns.distplot(scaled_time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time post Scaling', fontsize=14)
ax[1].set_xlim([min(scaled_time_val), max(scaled_time_val)])

plt.show()
```

## Importing Libraries for Model Training and Evaluation:

In this section, we import the necessary libraries for splitting the dataset, training various machine learning models, and evaluating their performance.

#### Libraries:
- **`train_test_split`**: 
  - From `sklearn.model_selection`, this function is used to split the dataset into training and testing sets.

- **`classification_report`**, **`confusion_matrix`**, **`precision_recall_curve`**, **`roc_auc_score`**, **`RocCurveDisplay`**:
  - From `sklearn.metrics`, these functions help in evaluating the performance of the models through various metrics and visualizations.

- **`RandomForestClassifier`**: 
  - From `sklearn.ensemble`, this is an ensemble learning method that builds multiple decision trees and merges them for improved accuracy.

- **`LogisticRegression`**: 
  - From `sklearn.linear_model`, this model is used for binary classification tasks.

- **`SVC`**: 
  - From `sklearn.svm`, this is a support vector classifier used for classification tasks.

- **`XGBClassifier`**: 
  - From the `xgboost` library, this is an implementation of gradient boosting that is particularly effective for structured data.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_auc_score, RocCurveDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X
```

```python
y
```

### Splitting the Dataset:

We use the `train_test_split` function from `sklearn.model_selection` to divide the dataset into training and testing sets. This is essential for evaluating model performance.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

### Training the Models:

In this section, we train multiple classification models using the training data. We will evaluate each model's performance on the test set.

```python
# Train Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
y_pred_proba_lr = lr.predict_proba(X_test)[:, 1]
```

```python
rf = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
y_pred_proba_rf = rf.predict_proba(X_test)[:, 1]
```

```python
svc = SVC(kernel='linear', probability=True, max_iter=1000, random_state=42)
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
y_pred_proba_svc = svc.predict_proba(X_test)[:, 1]
```

### Classification Reports:

After training the models, we evaluate their performance using classification reports. These reports provide key metrics such as precision, recall, and F1-score for each class.

```python
# Classification Reports
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("SVM Classification Report:\n", classification_report(y_test, y_pred_svc))
```

### Precision-Recall Curve:

The Precision-Recall curve is a valuable tool for assessing the performance of classifiers, especially in the context of imbalanced datasets. It visualizes the trade-off between precision and recall at various thresholds.

```python
# Precision-Recall Curve
precision_lr, recall_lr, _ = precision_recall_curve(y_test, y_pred_proba_lr)
precision_rf, recall_rf, _ = precision_recall_curve(y_test, y_pred_proba_rf)
precision_svc, recall_svc, _ = precision_recall_curve(y_test, y_pred_proba_svc)
```

```python
# Plotting Precision-Recall Curve
plt.plot(recall_lr, precision_lr, label='Logistic Regression')
plt.plot(recall_rf, precision_rf, label='Random Forest')
plt.plot(recall_svc, precision_svc, label='SVM')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

### Confusion Matrices:

Confusion matrices provide a detailed breakdown of the classification results, showing the true positives, true negatives, false positives, and false negatives. This is particularly useful for understanding model performance in imbalanced datasets.

```python
# Confusion Matrices
cm_lr = confusion_matrix(y_test, y_pred_lr)
cm_rf = confusion_matrix(y_test, y_pred_rf)
cm_svc = confusion_matrix(y_test, y_pred_svc)
```

```python
# Visualizing Confusion Matrices
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Logistic Regression Confusion Matrix')
plt.show()

sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Random Forest Confusion Matrix')
plt.show()

sns.heatmap(cm_svc, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('SVM Confusion Matrix')
plt.show()
```

## Model performance summary
While all models excelled at detecting non-fraudulent transactions (precision, recall, F1-score close to 1.00 for Class 0), they performed poorly for fraudulent transactions, with Class 1 metrics indicating a substantial drop in detection quality.

### Need for Handling Imbalanced Classes:
The significant performance gap between the two classes highlights the need for handling class imbalance. Without addressing this issue, models tend to favor the majority class (non-fraudulent) at the cost of missing out on the minority class (fraudulent), which is crucial in this scenario. Techniques such as SMOTE, ADASYN, or cost-sensitive learning can help improve the model's ability to correctly identify fraudulent transactions, improving overall robustness and reliability.
