### Splitting Data Using Stratified K-Fold

In this section, we use the **Stratified K-Fold** technique to split our dataset into training and testing sets while maintaining the original class distribution of the target variable. This is particularly important in the context of imbalanced datasets, such as in fraud detection, where one class (non-fraudulent transactions) may significantly outnumber the other class (fraudulent transactions).

#### Key Steps:

1. **Import StratifiedKFold**: We begin by importing the `StratifiedKFold` class from the `sklearn.model_selection` module. This class is designed to provide train/test indices to split data into train/test sets while preserving the percentage of samples for each class.

2. **Initialize StratifiedKFold**: We create an instance of `StratifiedKFold` with `n_splits=5`, which indicates that we want to split the data into 5 different sets for cross-validation. The `shuffle=False` parameter ensures that the data is not shuffled before splitting, preserving the order of samples.

3. **Split Data**: We use the `split` method of the `StratifiedKFold` instance to generate indices for the training and testing sets. During each iteration, it divides the dataset into training and testing subsets, ensuring that the class distribution in each fold is similar to that of the entire dataset.

4. **Store Indices**: The resulting indices are stored in `original_Xtrain`, `original_Xtest`, `original_ytrain`, and `original_ytest`, allowing us to use them in subsequent analysis and model training.

By employing this technique, we ensure that our machine learning models are trained and tested on data that reflects the true distribution of classes, which is crucial for accurately assessing their performance, especially in cases of imbalance.


```python
sss = StratifiedKFold(n_splits=5, shuffle=False)
```

```python
for train_index, test_index in sss.split(X, y):
    original_Xtrain, original_Xtest = X.iloc[train_index], X.iloc[test_index]
    original_ytrain, original_ytest = y.iloc[train_index], y.iloc[test_index]
```

## Random Under-Sampling: Theory and Mathematics

**Random Under-Sampling (RUS)** is a technique for handling imbalanced datasets by reducing the number of instances in the majority class to match the minority class. This ensures that the model does not become biased toward the majority class.

#### Mathematical Representation

1. **Let** \( D \) be the dataset:
   - \( N_m \) = instances in the majority class
   - \( N_n \) = instances in the minority class, where \( N_m \gg N_n \)

2. **Random Sampling**: 
   Select \( N_n \) instances from the majority class to form a balanced dataset:
   \[
   D_m' = \{x_1, x_2, \ldots, x_{N_n} | x_i \in D_m \}
   \]
   - \( D' = D_n \cup D_m' \), where \( D_n \) represents the minority class and \( D_m' \) represents the selected subset of the majority class.

3. **Class Distribution**:
   After under-sampling, the class probabilities are:
   \[
   P(Class_0) = P(Class_1) = 0.5
   \]

#### Research Paper References

1. **"A Systematic Review of Class Imbalance Learning"**  
   Reviews various imbalance handling methods including RUS, noting its effectiveness but cautioning about loss of information.

2. **"Impact of Random Under-Sampling on Imbalanced Data Classification"**  
   Discusses the bias-variance trade-off introduced by RUS, highlighting increased variance due to reduced dataset size.

3. **"Evaluating the Role of Data Sampling in Learning Classifiers"**  
   Compares RUS with other techniques, showing that while RUS reduces training time, it might omit important patterns from the majority class.

   - All models show **high recall** for classifying fraud cases (Class 1), with values ranging from 0.88 to 0.95. This means that most of the fraudulent transactions were correctly identified. However, **precision** is extremely low (between 0.04 and 0.07), indicating that many false positives occurredâ€”i.e., the models incorrectly flagged many non-fraudulent transactions as fraudulent.
   
2. **Impact of Random Under-Sampling**:
   - Random under-sampling helped in achieving **high recall** for the minority class (fraudulent transactions), which is often the goal when detecting rare events like fraud.
   - However, the **loss of information** from the majority class (non-fraud) due to under-sampling likely caused the models to perform poorly in terms of precision for the minority class. This happens because the model doesn't have enough examples of legitimate non-fraudulent transactions to learn from after under-sampling.

3. **Possible Overfitting**:
   - The models seem to have overfit to the **under-sampled training data**. This is indicated by the significant drop in precision when tested on the original dataset. The models learned the patterns well from the balanced training set, but these patterns did not generalize well to the original, highly imbalanced dataset, leading to a large number of false positives.

4. **Imbalanced Dataset Challenges**:
   - These results demonstrate the inherent challenge of handling **imbalanced datasets**. While under-sampling balances the classes and improves recall, it sacrifices the model's ability to generalize to the majority class, leading to a precision-recall trade-off that could make the model impractical in real-world applications.
